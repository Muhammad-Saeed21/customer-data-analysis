```python
# import all required libraries

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# For better chart style
sns.set_theme(style="whitegrid")
```

```python
# Load the CSV dataset

cust_dframe = pd.read_csv("customers-1000.csv")
print("First 5 rows: ")
cust_dframe.head()
```

```python
# Basic information about Dataset
print("Shape of Dataset: ")
# shows rows & columns
print(cust_dframe.shape)
```

```python
# Shows Names Of Columns
print("Column Names: ")
print(cust_dframe.columns)
```

```python
# Shows Summary Of Dataset
print("Summary: ")
print(cust_dframe.info())
```

```python
# Show summary statistics of numeric & datetime columns

cust_dframe.describe(include='all')
```

```python
# Check missing values count

cust_dframe.isna().sum()
```

```python
# Fill and Drop missing values
# Remove rows with All missing values
df_clean = cust_dframe.dropna()

# Replace missing values in Phone 2 with "Not Available"
df_clean['Phone 2'] = df_clean['Phone 2'].fillna("Not Available")
print("Show the first 5 rows after replacing missing values: ")
df_clean.head()
```

```python
# Convert Subscription Date to datetime

df_clean['Subscription Date'] = pd.to_datetime(df_clean['Subscription Date'])
# Extract Year & Month to create new columns for Year & Month
df_clean['Year'] = df_clean['Subscription Date'].dt.year
df_clean['Month'] = df_clean['Subscription Date'].dt.month
print("Display first 5 rows after adding year and month columns: ")
df_clean.head()
```

```python
# Groupby Country and count customers
country_count = df_clean['Country'].value_counts()
print("Display countries with customers counts: ")
country_count.head(30)
```

```python
# Bar chart of top 10 countries with most customers

plt.figure(figsize=(12,6))
sns.barplot(x=country_count.head(10).values,
            y=country_count.head(10).index)

plt.title("Top 10 Countries by Customer Count")
plt.xlabel("Count")
plt.ylabel("Country")
plt.grid()
plt.show()
```

```python
# Count customers by city

city_count = df_clean['City'].value_counts().head(10)
print("Show top 10 cities by customer count: ")
city_count
```

```python
# Top 10 Cities Bar Chart

plt.figure(figsize=(12,6))
sns.barplot(x=city_count.values, y=city_count.index)

plt.title("Top 10 Cities by Customer Count")
plt.xlabel("Count")
plt.ylabel("City")
plt.grid()
plt.show()
```

```python
# Count customers per subscription year
year_count = df_clean['Year'].value_counts().sort_index()
print("Shows customer count per subscription year: ")
year_count
```

```python
# Line chart of customer subscription trend

plt.figure(figsize=(10,5))
sns.lineplot(x=year_count.index, y=year_count.values, marker='>')

plt.title("Customer Subscription Per Year")
plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.grid()
plt.show()
```

```python
# Extract email domain for analysis

df_clean['Email_Domain'] = df_clean['Email'].str.split('@').str[1]
print("Display first 10 email domains: ")
df_clean['Email_Domain'].value_counts().head(10)
```

```python
# Bar Plot for top email domains

domain_count = df_clean['Email_Domain'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.barplot(x=domain_count.values, y=domain_count.index)

plt.title("Top 10 Email Domains")
plt.xlabel("Count")
plt.ylabel("Email Domain")
plt.show()
```
